```python
import SimpleITK as sitk
import numpy as np
import os
import matplotlib.pyplot as plt
from ipywidgets import interact, fixed

# Load an example image (Ensure this file exists)
image_path = "Dataset002_ThyroidSeg/imagesTr/case_001_0000.nrrd"

if not os.path.exists(image_path):
    raise FileNotFoundError(f"Image file not found: {image_path}")

# Read image using SimpleITK
image = sitk.ReadImage(image_path)
image_array = sitk.GetArrayFromImage(image)  # Convert to NumPy (Slices, Height, Width)

# Function to interactively view slices
def display_slice(slice_idx, image_array):
    plt.figure(figsize=(6, 6))
    plt.imshow(image_array[slice_idx, :, :], cmap="gray")
    plt.title(f"Slice {slice_idx}")
    plt.axis("off")
    plt.show()

# Interactive widget (Jupyter Notebook required)
interact(display_slice, slice_idx=(0, image_array.shape[0] - 1), image_array=fixed(image_array))

# Alternative: Show the middle slice if running as a script
mid_slice = image_array.shape[0] // 2
plt.figure(figsize=(6, 6))
plt.imshow(image_array[mid_slice, :, :], cmap="gray")
plt.title(f"Middle Slice {mid_slice}")
plt.axis("off")
plt.show()
```

```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt

# Load image
image_path = "Dataset002_ThyroidSeg/imagesTr/case_001_0000.nrrd"
logo = sitk.ReadImage(image_path)

# Convert to NumPy array
logo_array = sitk.GetArrayFromImage(logo)

# Select the middle slice
mid_slice = logo_array.shape[0] // 2
logo_2d = logo_array[mid_slice, :, :]  # Extract a 2D slice

# Brute force sub-sampling
logo_subsampled = logo_2d[::2, ::2]

# Get the sub-image containing the word "Simple"
simple = logo_2d[0:115, :]

# Get the sub-image containing "Simple" and flip it vertically
simple_flipped = simple[::-1, :]

# Create subplots
fig, axes = plt.subplots(4, 1, figsize=(6, 12))

# Display original 2D slice
axes[0].imshow(logo_2d, cmap="gray")
axes[0].set_title("Original Image (Middle Slice)")
axes[0].axis("off")

# Display sub-sampled image
axes[1].imshow(logo_subsampled, cmap="gray")
axes[1].set_title("Sub-sampled Image")
axes[1].axis("off")

# Display extracted "Simple" region
axes[2].imshow(simple, cmap="gray")
axes[2].set_title("Extracted 'Simple' Region")
axes[2].axis("off")

# Display flipped "Simple" region
axes[3].imshow(simple_flipped, cmap="gray")
axes[3].set_title("Flipped 'Simple' Region")
axes[3].axis("off")

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
#Image operations
#SimpleITK supports basic arithmetic operations between images while taking into account their meta-data. Images must physically overlap (pixel by pixel).
img_width = 128
img_height = 64
img1 = sitk.Image((img_width, img_height), sitk.sitkUInt8)
for i in range(img_width):
    img1[i, 1] = 5

img2 = sitk.Image(img1.GetSize(), sitk.sitkUInt8)
# img2.SetDirection([0,1,0.5,0.5])
img2.SetOrigin([0.000001, 0.000001])
for i in range(img_width):
    img2[i, 1] = 120
    img2[i, img_height // 2] = 60

# Add the two images (upper line has a value of 125, and lower of 60)
img3 = img1 + img2

plt.imshow(sitk.GetArrayViewFromImage(img3), cmap=plt.cm.Greys_r)
plt.axis("off");
```

```python
# operations can also be done in place
img1 += img2
plt.imshow(sitk.GetArrayViewFromImage(img1), cmap=plt.cm.Greys_r)
plt.axis("off");
#Comparative operators (>, >=, <, <=, ==) are also supported, returning binary images.
```
